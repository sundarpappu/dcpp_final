```python
#Wikipedia scrapping


import csv
import time
import json
import urllib
from urllib.request import urlopen, Request
import requests
from bs4 import BeautifulSoup
import pandas as pd


df = pd.read_excel("C:/Users/joyde/Downloads/Book9.xlsx")


artist = [] 
artist = df['Artist_temp']

for i in artist:
    url = 'https://en.wikipedia.org/wiki/'+i
    page = requests.get(url)


    soup = BeautifulSoup(page.text, 'lxml')
    soup

    headers = soup.find_all('th', class_='infobox-label')
    headers

    column_heads = []
    
    for i in headers:
        h = i.text
        column_heads.append(h)
        
    column_heads
    
    content = soup.find_all('td', class_='infobox-data')
    content 
    
    content_details = []
    
    for j in content:
        k = j.text
        content_details.append(k)
        
    content_details
    
    pair = zip(column_heads, content_details)
    artist_dict = dict(pair)
    
    print(artist_dict)
    
    df_artist = pd.DataFrame(artist_dict, [0])
    df = df.append(df_artist, ignore_index=False)
    print(df)
    
df.to_csv('C:/Users/joyde/OneDrive/Documents/wiki_scrape.csv')
```

```python
#Image extraction grammy

import re
import requests
from bs4 import BeautifulSoup

site = 'https://www.grammy.com/grammys/awards'

response = requests.get(site)

soup = BeautifulSoup(response.text, 'html.parser')
img_tags = soup.find_all('img')

urls = [img['src'] for img in img_tags]

list_img = []

for url in urls:
    filename = re.search(r'/([\w_-]+[.](jpg|gif|png))$', url)
    list_img.append(filename)
    if not filename:
         print("Regex didn't match with the url: {}".format(url))
         continue
    with open(filename.group(1), 'wb') as f:
        if 'http' not in url:
            # sometimes an image source can be relative 
            # if it is provide the base url which also happens 
            # to be the site variable atm. 
            url = '{}{}'.format(site, url)
        response = requests.get(url)
        f.write(response.content)
```
